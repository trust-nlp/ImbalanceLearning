以下代码只统计并可视化不同实体的数量

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

# # 2. 反转生成 id2label
id2label = {v: k for k, v in label2id.items()}

def count_entity_mentions(dataset, id2label):
    mention_counts = defaultdict(int)
    for example in dataset:
        for tag_id in example["tags"]:
            label_name = id2label[tag_id]
            # 跳过 "O" 标签
            if label_name == "O":
                continue
            if label_name.startswith("B-"):
                # 去掉前缀 B- 得到实体类型
                entity_type = label_name[2:]
                mention_counts[entity_type] += 1
    return mention_counts



# 5. 分别统计每个 split 的实体类型数
train_counts = count_entity_mentions(train_data, id2label)
valid_counts = count_entity_mentions(validation_data, id2label)
test_counts = count_entity_mentions(test_data, id2label)

print("数据集-实体类型计数：")
print("\nTrain：")
print(dict(train_counts))
print("\nValidation：")
print(dict(valid_counts))
print("\nTest：")
print(dict(test_counts))

# 1. 合并三个 split 的统计并按数量从多到少排序
merged = Counter(train_counts) + Counter(valid_counts) + Counter(test_counts)
entity_types, counts = zip(
    *sorted(merged.items(), key=lambda x: x[1], reverse=True)
)

# 2. 选择美观的 colormap
colors = plt.cm.viridis(np.linspace(0, 1, len(entity_types)))

# 3. 绘图
plt.figure(figsize=(10, 6))
plt.bar(entity_types, counts, color=colors)
plt.xlabel('Entity Type', fontsize=15)
plt.ylabel('Total Mentions', fontsize=15)
plt.title('tweetner7_2020 - Entity Counts Across Splits', fontsize=15)
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.show()
```

以下代码是 span level 的频率统计

```python
from collections import defaultdict

# 反转生成 id2label: 原数据集里是文字+id，要转换为label
id2label = {v: k for k, v in label2id.items()}

def count_tags_with_spans(dataset, id2label):
    """
    统计 dataset 中：
      - 非实体 token 'O' 的总数
      - 每个实体类型下，所有 B- 和 I- 开头标签的 token 数量
    返回一个 dict，比如 {'O': 12345, 'PER': 678, 'LOC': 345, ...}
    """
    counts = defaultdict(int)
    for example in dataset:
        for tag_id in example["tags"]:
            label = id2label[tag_id]
            
            if label == "O":
                # 统计所有 O
                counts["O"] += 1
            elif label.startswith(("B-", "I-")):
                # 统计实体 span 中所有的 B-XXX 和 I-XXX
                entity_type = label[2:]
                counts[entity_type] += 1
            # 其它标签（如果有），你也可以在这里根据需要统计
    
    return counts

# 假设你已定义好 id2label，并且 load_dataset 得到了 train_data, validation_data, test_data

train_counts = count_tags_with_spans(train_data, id2label)
valid_counts = count_tags_with_spans(validation_data, id2label)
test_counts  = count_tags_with_spans(test_data, id2label)

print("Train:", dict(train_counts))
print("Valid:", dict(valid_counts))
print("Test: ", dict(test_counts))



from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

# 1. 合并三个 split 的统计并按数量从多到少排序
merged = Counter(train_counts) + Counter(valid_counts) + Counter(test_counts)
entity_types, counts = zip(
    *sorted(merged.items(), key=lambda x: x[1], reverse=True)
)

# 2. 选择美观的 colormap
colors = plt.cm.viridis(np.linspace(0, 1, len(entity_types)))

# 3. 绘图
plt.figure(figsize=(10, 6))
plt.bar(entity_types, counts, color=colors)
plt.xlabel('Label Type', fontsize=15)
plt.ylabel('Total Mentions', fontsize=15)
plt.title('tweetner7_2020 - Label Counts Across Splits', fontsize=15)
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.show()

```
